# Question-1

Given a binary tree, your task is to find subtree with maximum sum in tree.

Examples:

Input1 :       

       1

     /   \

   2      3

  / \    / \

4   5  6   7

Output1 : 28

As all the tree elements are positive, the largest subtree sum is equal to sum of all tree elements.

Input2 :

       1

     /    \

  -2      3

  / \    /  \

4   5  -6   2

Output2 : 7

Subtree with largest sum is :

 -2

 / \

4   5

Also, entire tree sum is also 7.



```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def findMaxSubtreeSum(root):
    if not root:
        return 0

    max_sum = float('-inf')
    getSubtreeSum(root, max_sum)

    return max_sum

def getSubtreeSum(node, max_sum):
    if not node:
        return 0

    left_sum = getSubtreeSum(node.left, max_sum)
    right_sum = getSubtreeSum(node.right, max_sum)

    subtree_sum = node.val + left_sum + right_sum
    max_sum = max(max_sum, subtree_sum)

    return subtree_sum

# Create the binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Find the subtree with maximum sum
max_sum = findMaxSubtreeSum(root)

# Print the maximum sum
print("Maximum Subtree Sum:", max_sum)
```

# Question-2

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10



```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def constructBST(level_order):
    if not level_order:
        return None

    root = TreeNode(level_order[0])
    i = 1
    while i < len(level_order) and level_order[i] < root.val:
        i += 1

    root.left = constructBST(level_order[1:i])
    root.right = constructBST(level_order[i:])

    return root

def inorderTraversal(root):
    if not root:
        return []

    traversal = []
    traversal += inorderTraversal(root.left)
    traversal.append(root.val)
    traversal += inorderTraversal(root.right)

    return traversal

# Given level order traversal
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]

# Construct the BST
root = constructBST(level_order)

# Print the inorder traversal of the BST
print("Inorder Traversal:")
print(inorderTraversal(root))
```

# Question-3

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

```python
def canRepresentBST(arr):
    stack = []
    root = float('-inf')

    for element in arr:
        if element < root:
            return "No"

        while stack and element > stack[-1]:
            root = stack.pop()

        stack.append(element)

    return "Yes"

# Test case 1
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
print(canRepresentBST(arr1))  # Output: Yes

# Test case 2
arr2 = [11, 6, 13, 5, 12, 10]
print(canRepresentBST(arr2))  # Output: No
```
